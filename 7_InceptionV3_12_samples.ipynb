```python
import keras

from keras.applications.inception_v3 import InceptionV3
from keras.applications.inception_resnet_v2 import InceptionResNetV2

from keras.preprocessing.image import ImageDataGenerator
from keras.models import Model
from keras.models import load_model
from keras.layers import Dense, GlobalAveragePooling2D, Dropout
from keras.callbacks import ModelCheckpoint, EarlyStopping
#from keras.optimizers import SGD
from keras.utils.np_utils import to_categorical
from keras import backend as K

import h5py
import os
import shutil
import csv
import pickle

from PIL import Image

import numpy as np
from numpy import argmax

import seaborn as sn
import pandas as pd

import matplotlib
import matplotlib.pyplot as plt

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

versionstring = '23_InceptionV3_12_samples'
```

```python
img_height = 299
img_width = 299
CROP_LENGTH = 299
batch_size = 32
n_examples = 3
nr_of_output_classes = 69

# Training, test, and validation directories 
train_dir       = os.getcwd() + '/data/12_crops/train'
val_dir         = os.getcwd() + '/data/12_crops/valid'
test_dir        = os.getcwd() + '/data/12_crops/test'
```

```python
log_dir = os.getcwd() + '/logs/'
if not os.path.exists(log_dir):
    os.makedirs(log_dir)

save_dir = log_dir + versionstring + '/'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
    
model_dir = save_dir + 'models/'
if not os.path.exists(model_dir):
    os.makedirs(model_dir)

history_dir = save_dir + 'history/'
if not os.path.exists(history_dir):
    os.makedirs(history_dir)

print(log_dir)
print(save_dir)
print(model_dir)
print(history_dir)
```

```python
def random_crop(img):
    assert img.shape[2] == 3
    random = np.random.randint(1, 13)
    s = 299
    if random == 1:
        crop_img = img[0:s, s*0:s*1, :]
    elif random == 2:
        crop_img = img[0:s, s*1:s*2, :]
    elif random == 3:
        crop_img = img[0:s, s*2:s*3, :]
    elif random == 4:
        crop_img = img[0:s, s*3:s*4, :]
    elif random == 5:
        crop_img = img[0:s, s*4:s*5, :]
    elif random == 6:
        crop_img = img[0:s, s*5:s*6, :]
    elif random == 7:
        crop_img = img[s:s*2, s*0:s*1, :]
    elif random == 8:
        crop_img = img[s:s*2, s*1:s*2, :]
    elif random == 9:
        crop_img = img[s:s*2, s*2:s*3, :]
    elif random == 10:
        crop_img = img[s:s*2, s*3:s*4, :]
    elif random == 11:
        crop_img = img[s:s*2, s*4:s*5, :]
    elif random == 12:
        crop_img = img[s:s*2, s*5:s*6, :]
    
    return crop_img
```

```python
def crop_generator(batches):
    while True:
        batch_x, batch_y = next(batches)
        batch_crops = np.zeros((batch_x.shape[0], 299, 299, 3))
        for i in range(batch_x.shape[0]):
            batch_crops[i] = random_crop(batch_x[i])
        yield (batch_crops, batch_y)
```

```python
train_datagen = ImageDataGenerator(rescale=1./255)
valid_datagen  = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(598, 1794),
        color_mode="rgb",
        batch_size=batch_size,
        shuffle=True,
        class_mode='categorical')

train_crops = crop_generator(train_generator)

valid_generator = valid_datagen.flow_from_directory(
        val_dir,
        target_size=(299, 299),
        color_mode="rgb",
        batch_size=batch_size,
        shuffle=False,
        class_mode='categorical')

test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(299, 299),
        color_mode="rgb",
        batch_size=batch_size,
        shuffle=False,
        class_mode='categorical')
```

## InceptionV3

```python
base_model = InceptionV3(weights='imagenet', include_top=False, input_tensor=None,
               input_shape=(img_width,img_height,3))
```

```python
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(1024, activation='relu')(x)
x = Dropout(0.8)(x)
predictions = Dense(nr_of_output_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

#model.summary()
```

```python
for layer in model.layers:
   layer.trainable = True

#model.compile(optimizer=SGD(lr=0.001, momentum=0.9), loss='categorical_crossentropy', metrics=['acc'])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics = ['accuracy'])
```

```python
earlystopping = EarlyStopping(monitor='val_loss', min_delta=0, patience=20, verbose=1,
                              mode='auto', baseline=None, restore_best_weights=False)

checkpoint = ModelCheckpoint(filepath=model_dir+'Low_val_acc_{epoch:02d}.h5', monitor='val_loss', verbose=1,
                             save_best_only=True, save_weights_only=False, period=1)
```

```python
# This is used to continue training after pausing
# model = load_model('/22_InceptionV3_12_samples/models/Last_epoch.h5')
```

```python
history = model.fit_generator(
      train_crops,
      steps_per_epoch = train_generator.samples // batch_size,
      validation_data = valid_generator, 
      validation_steps = valid_generator.samples // batch_size,
      epochs=100, callbacks=[checkpoint])
```

```python
model.save(model_dir+'Last_epoch.h5')
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(acc) + 1)
```

## Test set performance

```python
model = load_model(model_dir+'Low_val_acc_52.h5')
```

```python
y_pred = model.predict_generator(test_generator, test_generator.samples//batch_size + 1,
                                 workers=1, use_multiprocessing=False, verbose = 1)
```

```python
y_est = np.argmax(y_pred,axis=1)
```

```python
cm = confusion_matrix(test_generator.classes,y_est)

mat = np.matrix(cm)
dataframe = pd.DataFrame(data=mat.astype(float))
csvfilename = save_dir+'CM.csv'
dataframe.to_csv(csvfilename, sep=',', header=False, float_format='%.f', index=False)
```

```python
print(accuracy_score(test_generator.classes,y_est))
```

```python
classes = sorted(os.listdir(test_dir))
classification_report = classification_report(test_generator.classes, y_est, target_names=classes, digits=4)
```
